```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
image_path = 'basin.png'
image = mpimg.imread(image_path)

fig, ax = plt.subplots()

ax.imshow(image, extent=[-1, 1, -1, 1])

plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.linspace(-1, 1, 5))
plt.yticks(np.linspace(-1, 1, 5))
plt.show()
```

```python
image_path = 'trajectory1.png'
image = mpimg.imread(image_path)

fig, ax = plt.subplots()

ax.imshow(image, extent=[0, 300, -1, 1])

plt.figure(figsize=(10, 10))
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.linspace(0, 300, 5))
plt.yticks(np.linspace(-1, 1, 5))
plt.show()
```
